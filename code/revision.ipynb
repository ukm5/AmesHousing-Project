```python
df_to_be_dummified = df_train[df_train.dtypes[df_train.dtypes == 'object'].keys()]
# df_to_be_dummified = pd.get_dummies(df_to_be_dummified, drop_first=True)
```

```python
X_columns = [i for i in df_train.columns if i not in {'id','pid','saleprice'}.union(set(df_to_be_dummified.columns))]
```

```python
X = df_train.loc[:,X_columns]
X = X.join(pd.get_dummies(df_to_be_dummified, drop_first=True))
X.head()
```

```python
y = df_train.loc[:,'saleprice']
y.head()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, PassiveAggressiveRegressor, ARDRegression, SGDRegressor, HuberRegressor, RANSACRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

# Linear Regression

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr_preds = lr.predict(X_test)
```

```python
lr.score(X_train, y_train)
```

```python
r2_score(y_test, lr_preds)
```

```python
lr_mse = mean_squared_error(y_test, lr_preds)
```

```python
lr_mse**0.5
```

# Ridge

```python
ridge = Ridge(alpha=100)
```

```python
ridge.fit(X_train, y_train)
```

```python
ridge_preds = ridge.predict(X_test)
```

```python
ridge.score(X_train, y_train)
```

```python
r2_score(y_test, ridge_preds)
```

```python
ridge_mse = mean_squared_error(y_test, ridge_preds)
```

```python
ridge_mse < lr_mse
```

```python
ridge_mse**0.5
```

# Lasso

```python
lasso = Lasso(alpha=100.0, max_iter=10000)
lasso.fit(X_train, y_train)
lasso_preds = lasso.predict(X_test)
lasso.score(X_train, y_train)
```

```python
r2_score(y_test, lasso_preds)
```

```python
lasso_mse = mean_squared_error(y_test, lasso_preds)
```

```python
lasso_mse < lr_mse
```

```python
lasso_mse**0.50
```
