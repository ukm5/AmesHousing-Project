# EDA and Cleaning

### Reading in the training dataset and looking for missing values and corrupt dtypes

```python
import pandas as pd
import numpy as np
import missingno as msno
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
pd.options.display.max_rows = 4000 
```

```python
data_train = pd.read_csv('./../datasets/train.csv')
```

```python
# Function to rename all columns to lowercase
def rename_to_snakecase(df_called):
    import pandas as pd
    df_called.rename(columns={item: '_'.join(item.lower().split()) for item in df_called.columns}, inplace = True)
    return df_called
```

```python
# Function called to rename columns
rename_to_snakecase(data_train);
```

```python
# Function to drop columns with mostly nan values based on the percentage specified
def function_to_drop_columns_heavy_in_nan(df_called,percent):
    import pandas as pd
    to_drop = list(df_called.columns[df_called.isna().mean() >= percent])
    df_called.drop(columns=to_drop, inplace=True)
    data_train.dropna(axis=0,how='all',inplace=True)
    return df_called
```

```python
# Dropping columns that have grater than or equal to 80% nan values
function_to_drop_columns_heavy_in_nan(data_train, 0.80);
```

```python
# Function to make linear combination of multiple columns based on specific coeffs
def summing_columns_linearly(df_called, column_names, column_coeffs, new_name_variable=None, drop_old_bool = False):
    import pandas as pd
    import numpy as np
    if (new_name_variable == None):
        new_name_variable = ' + '.join(column_names)
    df_called[new_name_variable] = np.dot((df_called.loc[:,column_names]),pd.Series(column_coeffs))
    return df_called
```

```python
# Adding the full and half baths to give total baths in basement and rest of house
summing_columns_linearly(data_train, column_names=['full_bath','half_bath'],column_coeffs=[1.0,0.5], new_name_variable='total_bath');
summing_columns_linearly(data_train, column_names=['bsmt_full_bath','bsmt_half_bath'],column_coeffs=[1.0,0.5], new_name_variable='bsmt_total_bath');
```

```python
cols_with_considerable_nan = list(data_train.isna().mean()[data_train.isna().mean() > 0.1].index)
data_train[cols_with_considerable_nan].dtypes
```

### Plot some variables to visualize their correlations

```python
data_train.corr();
```

```python
plt.figure(figsize=(10,15))
sns.heatmap(data_train.corr()[['saleprice']].sort_values(by='saleprice', ascending = False),annot=True, );
```

```python
def obtain_list_important_correlations(df_called, variable_dependent, corr_percent):
    mask = (df_called.corr()[[variable_dependent]].abs().sort_values(by=variable_dependent, ascending = False) > corr_percent)
    list_to_return = list(df_called.corr()[[variable_dependent]].abs().sort_values(by=variable_dependent, ascending = False)[mask].index)
    list_to_return.remove(variable_dependent)
    return list_to_return
```

```python
features = obtain_list_important_correlations(data_train, 'saleprice', 0.10)
features
```

```python
def scatter_plot(df_called, variable_dependent, list_of_features):
    for item in list_of_features:
        plt.figure(figsize=(7,7))
        plt.scatter(df_called[item], df_called[variable_dependent])
        plt.ylabel(variable_dependent)
        plt.xlabel(item)
        plt.title(variable_dependent +' vs '+ item );
```

```python
scatter_plot(data_train, 'saleprice', features);
```
