```python
# Function to make linear combination of multiple columns based on specific coeffs
def summing_columns_linearly(df_called, column_names, column_coeffs, new_name_variable=None, drop_old = False):
    import pandas as pd
    import numpy as np
    if (new_name_variable == None):
        new_name_variable = ' + '.join(column_names)
    df_called[new_name_variable] = np.dot((df_called.loc[:,column_names]),pd.Series(column_coeffs))
    if drop_old:
        df_called.drop(columns=column_names, inplace = drop_old)
    return df_called
```

```python
# Function to make scatter plots
def scatter_plot(df_called, variable_dependent, list_of_features):
    import matplotlib.pyplot as plt
    for item in list_of_features:
        plt.figure(figsize=(7,7))
        plt.scatter(df_called[item], df_called[variable_dependent])
        plt.ylabel(variable_dependent)
        plt.xlabel(item)
        plt.title(variable_dependent +' vs '+ item );
```

## Trial model check

```python
features = list(set(features))
```

```python
features += ['open_porch_sf','year_built','year_remod/add','full_bath','mas_vnr_area','fireplaces','alley_NONE', 'alley_Pave']
```

```python
len(features)
```

```python
X = df_clean[features]
y = df_clean['saleprice']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
X.shape
```

```python
y.shape
```

```python
lr = LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
```

```python
sc = StandardScaler()
```

```python
z_train = sc.fit_transform(X_train)
z_test = sc.transform(X_test)
```

```python
cross_val_score(lr, z_train,y_train, cv=5).mean()
```

```python
lr.fit(z_train, y_train)
```

```python
lr.score(z_train,y_train)
```

```python
lr.score(z_test,y_test)
```

```python
pred = lr.predict(z_test)
```

```python
r2_score(y_test, pred)
```
